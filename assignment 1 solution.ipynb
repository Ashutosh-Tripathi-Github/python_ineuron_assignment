## Question 2: Write a program which will find all such numbers which are divisible by 7 but are not a multiple of 5, between 2000 and 3200 (both included). The numbers obtained should be printed in a comma-separated sequence on a single line.

```python
start_limit=2000
end_limit=3200
final_list=[]
rem=start_limit%7
start_limit=start_limit+7-rem
final_list.append(start_limit)
while start_limit+7<=end_limit:
    start_limit=start_limit+7
    if start_limit%5!=0:
        final_list.append(start_limit)
print(final_list)
```

## Qusetion 1: Write a program which accepts a sequence of comma-separated numbers from console and generate a list.

```python
work_list=[]
result_list=[]
initial_input=input('Please enter sequence of comma-separated numbers : ')
work_list=initial_input.split(",")
for x in work_list:
    if x.isdigit():
        result_list.append(x)
print(result_list)
```

## Question 2: Create the below pattern using nested for loop in Python.
#### *
#### * *
#### * * *
#### * * * *
#### * * * * *
#### * * * *
#### * * *
#### * *
#### *

```python
total_rows=10
for i in range(0,total_rows):
    if i<=5:
        for j in range(0,i):
            print('* ',end='')
    else:
        for j in range(i,total_rows):
            print('* ',end='')
        
    print("\r") 
    
```

its citizens

SOCIALIST, SECULAR, DEMOCRATIC REPUBLIC
and to secure to all its citizens
